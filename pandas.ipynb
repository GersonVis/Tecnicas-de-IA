## Series

Los arreglos de tipo serie trabajan con dos arreglos, las series tienen un indice y un valor

```python
#importamos pandas
import pandas as pd
## importar numpy
import numpy as np
```

```python
## declarar una serie
## sintaxis: pd
serie=pd.Series([1,2,3,4,5])
serie
```

```python
## crear una serie con indices
serie_con_indices=pd.Series([1,2,3,4,5], index=[1,2,3,4,5])
serie_con_indices
```

```python
# serie con letras como indices
series_con_letras=pd.Series([1,2,3,4], index=["a", "b", "c", "d"])
series_con_letras
```

```python
## acceder a la información de la serie
series_con_letras["a"]
```

```python
## acceder a la informacion de serie de números
serie_con_indices[4]
```

```python
## especificar serie con tipo de dato entero
serie_con_tipo=pd.Series([1,2,4,5], dtype="int")
serie_con_tipo
```

```python
## especificar serie con tipo de dato float
serie_float=pd.Series(np.random.randint(1,10,10), dtype="float")
serie_float
```

```python
## colocar un nombre a la serie
serie_con_nombre=pd.Series(np.arange(1,5), name="numeros del uno al 5")
serie_con_nombre
```

## Crear series

Para crear una serie debemos pasar el argumento un diccionario con los datos y las claves para identificar cada valor
parametro lo pasaremos por argumento en el metodo .Series

```python
## creamos un diccionario, las claves serán los indices y su valor será el valor de la serie
coloresRGB={
    "blanco": "255, 255, 255",
    "negro": "0, 0, 0"
}
```

```python
## creamos una serie con pandas
serie_con_diccionario=pd.Series(coloresRGB, name="Serie de colores", dtype="string")
serie_con_diccionario
```

```python
## accedemos a la informacion de la serie
serie_con_diccionario["blanco"]
```

```python
## accedemos a otro valor del diccionario
serie_con_diccionario["negro"]
```

```python
## saber el tamaño de la serie
serie_con_diccionario.size
```

```python
## saber el tipo de dato de la serie
serie_con_diccionario.dtype
```

## Valores nulos

Nos ayudan a representar valores nulos, el tipo de dato por defecto en python es none
pero si usamos pandas o numpy el valor para representar nulos es NaN, non a number, no 
es un número, y se utiliza numpy para esta función, numpy,NaN

```python
## creando una serie con números nulos
serie_con_nulos=pd.Series([1,3,5,np.NAN])
serie_con_nulos
```

```python
## saber todos los valores nulos de una serie
serie_con_nulos.isnull()
```

```python
## obtener los valores nulos
serie_con_nulos[serie_con_nulos.isnull()]
```

```python
## crear un arreglo
array=np.random.rand(3)
array
```

```python
## invertir un array
array[::-1].sort()
array
```

```python
## valores que no son nulos
serie_con_nulos.notnull()
```

```python
## obtener todos los valores que no son nulos
serie_con_nulos[serie_con_nulos.notnull()]
```

## Dataframe

un dataframe es muy parecido a una hoja de datos de excel, con columnas y filas

```python
## crear un diccionario cada key del diccionario será el nombre la columa
## dentro de los [] ira la información contenida en esa columna
## diccionario alumnos:
diccionario_alumnos={
    "nombre":["gerson", "german", "daniel"],
    "edad":[23, 19,20],
    "e-mail":["gerson@escuela.com", "german@escuela.com", "daniel@escuela.com"],
}
```

```python
## crear un dataFrame con el diccionario
dataframe_alumnos=pd.DataFrame(diccionario_alumnos)
dataframe_alumnos
```

```python
## acceder a los datos de un dataframe
## acceder a al acolumna ombre  de alumnos
dataframe_alumnos.nombre
```

```python
## se puede acceder a las columnas atravez de corchetes
## nota: es mejor acceder de esta manera a la informacion de las columnas
dataframe_alumnos['nombre']
```

```python
## acceder a un registro de una columna
dataframe_alumnos['nombre'][1]
```

```python
## acceder a una fila de un dataframe
dataframe_alumnos.values[0]
```

```python
## acceder a los valores contenidos en un dataframe
dataframe_alumnos.values
```

```python
## acceder al nombre de la s columnas de un dataframe
dataframe_alumnos.columns
```

## Columnas del dataframe

Administrar columnas del dataframe, eliminar, crear y modificar, cada columna es en realidad una serie

```python
## añadir una nueva columna
## crear una serie
color_favorito=pd.Series(["rojo", "azul", np.NAN], index=["a","b", "c"])
dataframe_alumnos["color favorito"]=color_favorito
dataframe_alumnos
```

```python
## visualizamos NAN por que tiene diferentes indices
## renombrar indices de color favorito
color_favorito=pd.Series(["rojo", "azul", np.NAN])
dataframe_alumnos["color favorito"]=color_favorito
dataframe_alumnos
```

```python
## mover los indices
color_favorito=pd.Series(["rojo", "azul", np.NAN], index=[1,2,3])
dataframe_alumnos["color favorito"]=color_favorito
dataframe_alumnos
```

```python
## renombrar una columna, el resultado no se aplica solo retorna un dataframe con los cambios
dataframe_alumnos.rename(
   columns={"color favorito":"colorfavorito"}
)
```

```python
## hacer permanente el renombre de las columnas
dataframe_alumnos=dataframe_alumnos.rename(
   columns={"color favorito":"colorfavorito"}
)
dataframe_alumnos
```

```python
## eliminar una columna
del dataframe_alumnos['colorfavorito']
```

## Ordenamiento

```python
## cargar datos de csv 
dataframe=pd.read_csv("informacion.csv", index_col="id")
dataframe[:5]
```

```python
## limpiar datos, cambiar los nan de nombre por sin nombre
dataframe=dataframe.fillna({"nombre":"sin nombre"})
dataframe
```

```python
## obtener la eda más joven de canada
dataframe[dataframe["pais"]=="Canada"].head(1)
```

```python
## obtener los 5 hombre más viejos de alemania
dataframe[(dataframe["pais"]=="Germany") & (dataframe["genero"]=="male")].sort_values("edad", ascending=False).values[:2]
```

## Busqueda por rangos

```python
## personas que tienen entre 40 y 50 años
dataframe[(dataframe["edad"]>=40)&(dataframe["edad"]<=50)]
```

```python
## usar la sentencia between
dataframe[dataframe["edad"].between(40,50)].values[:4]
```

```python
## obtener todos las personas que sean mayores de 18 años de los paises de France, Germany e Italy
dataframe[(dataframe["edad"]>=18)&((dataframe["pais"]=="Germany")|(dataframe["pais"]=="Netherlands")|(dataframe["pais"]=="Ireland"))]
```
