## Agregación de datos por categoría

```python
## importacion de librerias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
## opciones posibles en un dataset
gender = ["male", "female"]
income = ["poor", "middle class", "rich"]
```

```python
## cantidad de registros en el dummy
n=500
```

```python
## crear posibilidades aleatorias de genero
gender_data=[np.random.choice(gender) for pos in range(n)]
gender_data
```

```python
## nivel economico
income_data=[np.random.choice(income) for pos in range(n)]
income_data
```

```python
## crear diferentes datas para diferentes caracteristicas
## distribución normal N(m,z): m+n*random m=media, n=desviación tipica z=distribución de números aleatorios, normal 0
```

```python
height=160+30*np.random.rand(n)
height
```

```python
## para pesos
weight=70+40*np.random.rand(n)
weight
```

```python
## para edades
age=60+40+np.random.rand(n)
age
```

```python
## cantidad de ingresos con una media de 8000 pesos y una desviación estandar de 5000 pesos
income=8000+5000*np.random.rand(n)
income
```

```python
## crear edades con una media de 62 años y una desviacion de 30 años
age_data=62+20*np.random.rand(n)
age_data
```

```python
dummy=pd.DataFrame({
    "gender": gender_data,
    "age": age_data,
    "economic status": income_data,
    "height": height,
    "weight": weight,
    "income": income 
})
dummy.head()
```

```python
## agrupación de los datos por categoría
## agrupar los datos por genero
grouped_gender=dummy.groupby("gender")
```

```python
## ver los grupos de una agrupacion
grouped_gender.groups.keys()
```

```python
## recorrer los datos de una agrupacion
for group, data in grouped_gender:
    print(group)
    print(data)
```

```python
## obtener un grupo de un dataframe
grouped_gender.groups["male"]
```

```python
## doble agrupación, se hace agrupación por genero y por cada genero se hace una agrupacion para estatus economico
double_group=dummy.groupby(["gender", "economic status"])
len(double_group.groups)
```

```python
# visualizar grupos
double_group.groups
```

```python
## visualizamos los nombres de los grupos
double_group.groups.keys()
```

```python
## accceder a una doble agrupación
double_group.groups[('female', "poor")]
```

## Agregación de datos

```python
## suma de datos agrupados
double_group.sum()
```

```python
## promedio de datos agrupados
double_group.mean()
```

```python
## saber cuantos elementos hay en cada una de las combinaciones
double_group.size()
```

```python
## informacion de los datos agrupados
double_group.describe()
```

```python
## grupo de ingresos
group_income=dummy.groupby("income")
```

```python
## operaciones por agregacion, cada funcion se hace con cada columna
double_group.aggregate({
    "income": np.sum,
    "height": np.mean
})
```

```python
## aplicar nuestras propias funciones de agregación con lambda
double_group.aggregate({
    "height": lambda x: (0 if np.sum(x)<100 else 1)
})
```

## Filtrado, transformación y otras operaciones

## Transformacion de variables

```python
## tipificar los datos, saber que tan probables son
dummy["height"]=(dummy[["height"]].values-dummy[["height"]].values.mean())/dummy[["height"]].values.std()
```

```python
dummy
```

```python
## tipificar los datos de una agrupación
z_group=double_group.transform(lambda x: (x-x.mean())/x.std())
```

```python
plt.hist(z_group["age"])
```

## Dividir un conjunto de entrenamiento y un conjunto de testing

```python
## cargar los datos del dataset
data=pd.read_csv("archivos/churn model.txt")
data.describe()
```

## dividir utilizando la distribución normal

```python
## longitud del dataset
n=len(data)
## crear un arreglo de la misma cantidad del dataset de distribución normal
distribucion_normal=np.random.randn(n)
## tipificar distribución
tipificar_distribucion=distribucion_normal/distribucion_normal.max()
plt.hist(tipificar_distribucion)
```

```python
## obtener los que sean menores que 0.8
valores_verdaderos=int(len(data)*.25)
condicion=np.ones(valores_verdaderos)
condicion_falsa=np.zeros(len(data)-valores_verdaderos)
total=np.concatenate((condicion,condicion_falsa))
plt.hist(total.astype(int))
```

```python
data.size
```

```python
## convertir a booleano
conversor=np.vectorize(lambda x: x==1)
total=conversor(total)
```

```python
training=data[total]
test=data[~total]
print(len(training))
print(len(test))
```

```python
## librería para obtener muestras de datos
from sklearn.model_selection import train_test_split
```

```python
## obtener los datos de training y de testing usando la librería
train, test =train_test_split(data, test_size=0.2)
print(len(train))
print(len(test))
```

```python
## usando una función shuffle
## import random de sklearn
import sklearn
```

```python
## mezclar las filas aleatoriamente, y almacenamos el resutlado
data=sklearn.utils.shuffle(data)
data
```

```python
## dividir data
n=int(len(data))
train_size=int(n*.25)
train=data[:train_size]
test=data[train_size:]
print(len(train))
print(len(test))
print(len(data))
print(len(train)+len(test))
```
