## Agregación de datos por categoría

```python
## importacion de librerias
import numpy as np
import pandas as pd
```

```python
## opciones posibles en un dataset
gender = ["male", "female"]
income = ["poor", "middle class", "rich"]
```

```python
## cantidad de registros en el dummy
n=500
```

```python
## crear posibilidades aleatorias de genero
gender_data=[np.random.choice(gender) for pos in range(n)]
gender_data
```

```python
## nivel economico
income_data=[np.random.choice(income) for pos in range(n)]
income_data
```

```python
## crear diferentes datas para diferentes caracteristicas
## distribución normal N(m,z): m+n*random m=media, n=desviación tipica z=distribución de números aleatorios, normal 0
```

```python
height=160+30*np.random.rand(n)
height
```

```python
## para pesos
weight=70+40*np.random.rand(n)
weight
```

```python
## para edades
age=60+40+np.random.rand(n)
age
```

```python
## cantidad de ingresos con una media de 8000 pesos y una desviación estandar de 5000 pesos
income=8000+5000*np.random.rand(n)
income
```

```python
dummy=pd.DataFrame({
    "gender": gender_data,
    "economic status": income_data,
    "height": height,
    "weight": weight,
    "income": income 
})
dummy.head()
```

```python
## agrupación de los datos por categoría
## agrupar los datos por genero
grouped_gender=dummy.groupby("gender")
```

```python
## ver los grupos de una agrupacion
grouped_gender.groups.keys()
```

```python
## recorrer los datos de una agrupacion
for group, data in grouped_gender:
    print(group)
    print(data)
```

```python
## obtener un grupo de un dataframe
grouped_gender.groups["male"]
```

```python
## doble agrupación, se hace agrupación por genero y por cada genero se hace una agrupacion para estatus economico
double_group=dummy.groupby(["gender", "economic status"])
len(double_group.groups)
```

```python
# visualizar grupos
double_group.groups
```

```python
## visualizamos los nombres de los grupos
double_group.groups.keys()
```

```python
## accceder a una doble agrupación
double_group.groups[('female', "poor")]
```

## Agregación de datos

```python
## suma de datos agrupados
double_group.sum()
```

```python
## promedio de datos agrupados
double_group.mean()
```

```python
## saber cuantos elementos hay en cada una de las combinaciones
double_group.size()
```

```python
## informacion de los datos agrupados
double_group.describe()
```

```python
## grupo de ingresos
group_income=dummy.groupby("income")
```

```python
## operaciones por agregacion, cada funcion se hace con cada columna
double_group.aggregate({
    "income": np.sum,
    "height": np.mean
})
```

```python
## aplicar nuestras propias funciones de agregación con lambda
double_group.aggregate({
    "height": lambda x: (0 if np.sum(x)<100 else 1)
})
```

## Filtrado, transformación y otras operaciones

```python
## para filtrar se pasa una funcion lambda que comprueba cada uno de los datos
## devuelve el elemento que dicha comprobación sea verdadera
## filtrar las estaturas de los grupos de acuerdo a la función
## todos los registros con estaturas mayores a 1.5
double_group["height"].filter(lambda h: np.mean(h)>1.5)
```

```python
dummy["height"].filter(lambda x: np.mean(x)>1)
```
