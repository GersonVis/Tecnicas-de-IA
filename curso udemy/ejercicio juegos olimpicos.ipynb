```python
## especificar que datos quedaran a la izquierda y que datos quedaran a la derecha
## merge medallas y paises
medals_country=pd.merge(right=medals, left=country,
                        left_on="Athlete", right_on="Athlete")
medals_country.shape
```

```python
## unimos los dos datasets 
medals_country_not_dp=pd.merge(right=medals, left=country_not_dp,
                               right_on="Athlete", left_on="Athlete")
medals_country_not_dp.head(5)
```

```python
country_not_dp=country_not_dp.drop_duplicates("Athlete")
```

```python
## obtener 6 nombres aleatorios del dataset countrys
out_athletes=np.random.choice(country_not_dp["Athlete"], size=6, replace=False)
out_athletes
```

```python
dataframe_without = lambda dataFrame, out_items, column_name: ~dataFrame[column_name].isin(out_items)
```

```python
country_not_athletes=country_not_dp[(dataframe_without(country_not_dp,out_athletes,"Athlete"))
                                    &(country_not_dp["Athlete"]!="Michael Phelps")]
country_not_athletes
```

```python
medals_not_athletes=medals_country_not_dp[(dataframe_without(medals_country_not_dp,out_athletes,"Athlete"))
                                    &(medals_country_not_dp["Athlete"]!="Michael Phelps")]
medals_not_athletes
```

```python
sports_not_athletes=sports_not_dp[(dataframe_without(sports_not_dp,out_athletes,"Athlete"))
                                    &(sports_not_dp["Athlete"]!="Michael Phelps")]
sports_not_athletes
```

## ejemplo de joins con python

```python
dataA=pd.DataFrame({
    "nombre":["gerson", "jerman", "juan", "pablo", "fernando"],
    "nacionalidad": ["mexico", "usa", "puerto rico", "mexico", "uruguay"],
    "deporte": ["futbol", "basquetbol", "boleivol", "basquet", "basquet"]
})
```

```python
dataB=pd.DataFrame({
    "pais": ["usa", "mexico", "puerto rico", "venezuela", "francia"],
    "comida tipica": ["hamburguesa", "pozole", "palomas", "pechuga", "espagueti"]
})
dataB
```

```python
dataC=pd.DataFrame({
    "platillo": ["hamburguesa", "pozole", "palomas", "pechuga", "espagueti", "caldo de maricos"],
    "precio":  100+50*np.random.randn(6)
})
```

```python
inner_join=pd.merge(left=dataA, right=dataB, left_on="nacionalidad", right_on="pais")
inner_join
```

```python
inner_join=pd.merge(left=dataA, right=dataB, left_on="nacionalidad", right_on="pais", how="inner")
inner_join
```

```python
## hacer un left join tener los daos que no estÃ¡n en la interseccion de a y b
## personas sin nacionalidad
left_join=pd.merge(left=dataA, right=dataB, left_on="nacionalidad", right_on="pais", how="left")
left_join[pd.isna(left_join["pais"])]
```

```python
right_join=pd.merge(left=dataA, right=dataB, left_on="nacionalidad", right_on="pais", how="right")
right_join
```

```python
## aÃ±adir un registro
dataB.append({
    "pais": "cuba",
    "comida tipica": "arroz"
    }, ignore_index=True
    )
```

```python
## outer todo de todo
outer_join=pd.merge(left=dataA, right=dataB, left_on="nacionalidad", right_on="pais", how="outer")
outer_join[(pd.isna(outer_join["nombre"]))]
```
