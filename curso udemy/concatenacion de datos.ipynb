```python
## en python tenemos dos tipos de ejes
## axis=0 el eje horizontal axis=1 es vertical
# concatenar informacipon
## se concatena con el primer dataset introducido, el orden influye el orden de los datos
wine_data=pd.concat([white_wine, red_wine], axis=0)
```

```python
## tama√±o del dataset
wine_data.shape
```

```python
## ejercicios de dataset
data1=wine_data.head(10)
data2=wine_data[300:310]
data3=wine_data.tail(10)
```

```python
## concatenar datasets
wine_scramble=pd.concat([data1,data2,data3], axis=0)
wine_scramble
```

```python
## verificar que el orden afecta el resultado
other_scramble=pd.concat([data3, data1, data2], axis=0)
other_scramble
```

## carga de cientos de datos distribuidos

```python
## cargar el primer csv para ver su contenido y as√≠ proceder a ver los demas csv
## matriz esparce, la mayor√≠a de los datos no existen, sparce
primer_archivo=pd.read_csv("archivos/distributed-data/001.csv")
primer_archivo
```

```python
primer_archivo.shape
```

```python
## importar el fichero primero
## recorrer los ficheros
## debe haber consistencia de datos
## cada uno de ellos debe appendisarce al primero
## repetir el bucle hasta que no existan ficheros
## ruta de los archivos
file_path="archivos/distributed-data/"
data=pd.read_csv(file_path+"001.csv")
final_lenght=0
for momenclatura in range(1, 333):
    ruta=file_path+"%03i.csv"%momenclatura
    print(ruta)
    nuevo_cargado=pd.read_csv(ruta)
    final_lenght+=nuevo_cargado.shape[0]
    data=pd.concat([data, nuevo_cargado])
print("%s %s"%(final_lenght, data.shape[0]))
```
